```python
import pickle
import cv2
import matplotlib

import glob
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
output_grid, output_n, output_grids = pickle.load(open( "./output/new2_densenet_allimages_down64occsensitivity_female_all.p", "rb" ))
```

```python
output_n
```

```python
step = 4
n_slices = int(64/4)
i = 0
n = 0
data_0 = (output_grid[0]/output_n[0])
data_1 = (output_grid[1]/output_n[1])
data_2 = (output_grid[1]/output_n[1]-output_grid[0]/output_n[0])
slice = 0
fig, ax = plt.subplots(3, n_slices, figsize=[18, 1.2*3])
for _ in range(n_slices):
  #tmp_data = cv2.applyColorMap(cv2.cvtColor(data[:,:,slice], cv2.COLOR_GRAY2BGR), colormap)
  ax[0,n].imshow(data_0[:,:,slice], norm=matplotlib.colors.Normalize(vmin=np.min(data_0), vmax=np.max(data_0), clip=False))
  ax[0,n].set_xticks([])
  ax[0,n].set_yticks([])
  if i == 0:
    ax[0,n].set_title(str(slice), color='black')
  else:
    ax[0,n].set_title('', color='black')
  ax[1,n].imshow(data_1[:,:,slice], norm=matplotlib.colors.Normalize(vmin=np.min(data_1), vmax=np.max(data_1), clip=False))
  ax[1,n].set_xticks([])
  ax[1,n].set_yticks([])
  ax[1,n].set_title('', color='black')
  ax[2,n].imshow(data_2[:,:,slice], norm=matplotlib.colors.Normalize(vmin=np.min(data_2), vmax=np.max(data_2), clip=False))
  ax[2,n].set_xticks([])
  ax[2,n].set_yticks([])
  ax[2,n].set_title('', color='black')
  n += 1
  slice += step
```
